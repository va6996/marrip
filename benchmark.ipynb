```python
def comparison(d1, d2):
  res = []
  
  for e1,e2 in zip(d1,d2):
    e1d = e1.split(",")
    e2d = e2.split(",")
    tot1 = int(e1d[1])
    tot2 = int(e2d[1])
    miss1 = int(e1d[3])
    miss2 = int(e2d[3])
    r1 = miss1/tot1
    r2 = miss2/tot2
    # % fewer misses
    percent = (r1-r2)/r1*100
    res.append(percent)
  return res
```
